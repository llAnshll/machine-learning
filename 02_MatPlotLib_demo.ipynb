```python
import numpy as np
# Numpy :- import numpy liabray 
import pandas as pd
# Pandas :- import pandas liabray 
import matplotlib.pyplot as plt
# MatPlotLib :- import matplotlib liabray 
```

```python
x=np.array([10,25])
y=np.array([0,350])
#plot :-draw points (markers) in a diagram
plt.plot(x,y)
#show :-starts an event loop
plt.show()
```

```python

plt.plot(x,y,'o')
plt.show()
# Marker Type
#'o' 	Circle 	
#'*' 	Star 	
#'.' 	Point 	
#',' 	Pixel 	
#'x' 	X 	
#'X' 	X (filled) 	
#'+' 	Plus 	
#'P' 	Plus (filled) 	
#'s' 	Square 	
#'D' 	Diamond 	
#'d' 	Diamond (thin) 	
#'p' 	Pentagon 	
#'H' 	Hexagon 	
#'h' 	Hexagon 	
#'v' 	Triangle Down 	
#'^' 	Triangle Up 	
#'<' 	Triangle Left 	
#'>' 	Triangle Right 	
#'1' 	Tri Down 	
#'2' 	Tri Up 	
#'3' 	Tri Left 	
#'4' 	Tri Right 	
#'|' 	Vline 	
#'_' 	Hline
#Line Type
#'-' 	Solid line 	
#':' 	Dotted line 	
#'--' 	Dashed line 	
#'-.' 	Dashed/dotted line
#Colors
#'r' 	Red 	
#'g' 	Green 	
#'b' 	Blue 	
#'c' 	Cyan 	
#'m' 	Magenta 	
#'y' 	Yellow 	
#'k' 	Black 	
#'w' 	White
```

```python
x1=np.array([10,20,15,25])
y1=np.array([0,12,10,35])
plt.plot(x1,y1)
plt.show()
```

```python
x1=np.array([10,20,15,25])
y1=np.array([0,12,10,35])
plt.plot(x1,y1,'--')
plt.show()
```

```python
x1=np.array([10,20,15,25])
y1=np.array([0,8,12,22])
plt.plot(x1,y1,'r')
plt.show()
```

```python
plt.plot(x1,y1,marker='*',ms=10,mec='r',mfc='y',ls=':')
plt.show()
# ms= marker size
# mec= marker edge color
# mfc= marker face color
# ls = lines style
```

```python
plt.plot(x1,y1,linestyle='dotted',color='r',linewidth='10')
plt.show()
```

```python
plt.plot(x1,y1,x,y)
plt.show()
```

```python
font={'color':'green','size':'20'}
plt.title("line chart",fontdict=font,loc="left")
plt.xlabel("number1")
plt.ylabel("number2")
plt.plot(x1,y1)
plt.grid(axis='x',color='g',ls='--',lw='0.5')
plt.grid(axis='y',color='r',ls='--',lw='0.5')
```

```python
plt.subplot(1,2,1)
plt.title("p1")
plt.plot(x,y)
plt.subplot(1,2,2)
plt.title("p2")
plt.plot(x1,y1)
plt.suptitle("super title")
```

```python
x=np.array([10,25])
y=np.array([5,30])
x1=np.array([20,35])
y1=np.array([15,50])
sizes=np.array([10,50])
plt.scatter(x1,y1,color='red')
plt.scatter(x,y,color='green',s=sizes, alpha=0.5) 
#alpha is for transperancy
```

```python
plt.bar(x1,y1,color='red',width=2)
plt.show()
```

```python
plt.barh(x1,y1,color='r',height=2)
plt.show()
```

```python
plt.pie(x1)
plt.show()
```

```python
track_stud=np.array([20,30,40,20,10])
track_name=["Data","Mobile","Web","IoT","CyberSecurity"]
ex=[0.2,0,0,0,0]
col=["red","pink",'m','g','b']
plt.pie(track_stud,labels=track_name,startangle=90,explode=ex,shadow=True,colors=col)
plt.legend(title="TRACKS")
plt.show()
```
